```python
# Assuming 'df' as the dataset loaded
summary_statistics = df.describe(include='all')
print(summary_statistics)
```

```python
#custom_palette = sns.color_palette("colorblind", n_colors=2)
custom_palette = sns.color_palette(["#FFA500", "#4169E1"])
categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'poutcome']
for feature in categorical_features:
    plt.figure(figsize=(15, 6))
    sns.countplot(x=feature, hue='y', data=df,palette=custom_palette)
    plt.title(f'Distribution of {feature} with respect to y')
    plt.show()
```

```python
numerical_features = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']

for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.histplot(x=feature, hue='y', data=df, kde=True)
    plt.title(f'Distribution of {feature} with respect to y')
    plt.show()
```

```python
sns.pairplot(df, hue='y')
plt.title('Pairplot of Numerical Features')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Setting the style to whitegrid for gridlines
sns.set(style="whitegrid")

# Plotting the distribution of the target variable 'y'
plt.figure(figsize=(8, 6))
sns.countplot(x='y', data=df, palette='viridis')
plt.xlabel('Subscription Outcome')
plt.ylabel('Count')
plt.title('Distribution of Subscription Outcome')
plt.show()
```

```python
df['y'] = [0 if x == 'no' else 1 for x in df['y']]
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
